**Commercial Recommender System: YouTube**

I use YouTube most of the time and it is one type of commercial recomender system. YouTube is where I go to listen to music watch movie clips and reviews learn things from videos and stay up to date on topics that I am interested in. Since YouTube has a lot of content it has to use systems that recommend things to each user.

The way YouTube recommends things seems to be a mix of looking at what other people like me're watching looking at the content of the videos themselves and using machine learning. When YouTube looks at what other people like me're watching it is called collaborative filtering. For example, if people who watch videos about programming also watch videos about data science YouTube might recommend those data science videos to me too. When YouTube looks at the content of the videos it is called content-based filtering. This means it looks at things like categories and keywords. If I watch a lot of Python programming videos YouTube will probably recommend videos about Python.

YouTube also uses machine learning to look at what I do on the site. This includes what I watch what I search for how long I watch things what I like, what I subscribe to, what I comment on and what I click on. YouTube is always updating what it recommends to me based on what I have done. This means it can quickly change when my interests change.

I think YouTube does a job of recommending things to me. When I listen to music it usually suggests songs and artists that I like. When I watch videos it recommends videos that help me keep learning. When I watch movie trailers it recommends videos. I like that YouTube helps me find things to watch.

YouTube is not perfect though. Sometimes, it recommends the kinds of videos over and over. This is called a filter bubble. It can stop me from seeing things that I might like. Sometimes YouTube gets too focused on what I watched and usually recommends things that are not really what I am interested in anymore. With these problems I think YouTubes recommendations are usually really good.

Overall, YouTube shows how looking at what other people're watching looking at the content of videos and using machine learning can all work together to make a system that recommends things to millions of people every day. I use YouTube. It helps me find new things to watch. YouTube is an example of how these systems can work together to help people, like me find new things.

You can observe the same kind of occurrence happening on Amazon. Some sellers make accounts and pay people to provide their products good reviews. Other sellers work together to give each other good reviews. Since Amazons Recommender System relies on these ratings and reviews, the fake ones make products look really popular. This affects what people decide to buy often confusing them and making it difficult for them to select the best option. As a result, people see products that look highly rated. The ratings are not really about how good the products are.

To stop this kind of abuse I would make a Recommender System that cares more about what people do. For instance, only people who have bought a product should be able to give reviews that affect the recommendation scores. The system can use machine learning to find patterns, like a lot of reviews coming from the computer or a lot of reviews happening in a short amount of time. It can also find groups of accounts that always review the products together. Reviews from accounts can count less than reviews from people who have been using the system for a while because these people have a history of leaving reviews.

Another good idea is to use things than ratings to make recommendations. For example, we can look at how people return products or if they buy something again. We can also look at if they keep coming to the site or if they are happy with their purchases. These things give us an idea of how good a product is. By doing these things the Recommender System would be harder to trick. It would still let real customers help make recommendations.
